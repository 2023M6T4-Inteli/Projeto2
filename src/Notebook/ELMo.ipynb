# Modelo de vetorização - ELMo

ELMo corresponde a um modelo de vetorização que leva em consideração o contexto de uma palavra. O ELMo foi desenvolvido a fim de melhorar a indicação da próxima palavra que se encaixa ao contexto, como, por exemplo, a sugestão de palavras ao escrever um e-mail. (Wei, 2020)

Já o modelo pré-treinado em português foi desenvolvido pela Universidade Federal de Goiás (UFG). Disponível em: `https://allenai.org/allennlp/software/elmo`

## Teste

```python
options_file = 'https://s3-us-west-2.amazonaws.com/allennlp/models/elmo/contributed/pt/brwac/options.json'
weight_file = "/content/drive/MyDrive/Colab Notebooks/SI-MOD6/entrega_spt4/elmo_pt_weights_dgx1 (1).hdf5"
elmo = Elmo(options_file, weight_file, num_output_representations=1, dropout=0)

# Convert sentences to character IDs
sentences = [pipeline['texto']]
character_ids = batch_to_ids(sentences)

# Compute ELMo embeddings
embeddings = elmo(character_ids)['elmo_representations'][0]

# Convert embeddings to NumPy array
embeddings = embeddings.detach().numpy()

```

## Função


```python
def elmo():

  options_file = 'https://s3-us-west-2.amazonaws.com/allennlp/models/elmo/contributed/pt/brwac/options.json'
  weight_file = "/content/drive/MyDrive/Colab Notebooks/SI-MOD6/entrega_spt4/elmo_pt_weights_dgx1 (1).hdf5"
  elmo = Elmo(options_file, weight_file, num_output_representations=1, dropout=0)

  # conversão de sentenças em ids
  sentences = [pipeline['texto']]
  character_ids = batch_to_ids(sentences)

  # computa ids em vetores
  embeddings = elmo(character_ids)['elmo_representations'][0]

  # converte em vetores e establiza em array
  embeddings = embeddings.detach().numpy()
```

# Rede neural simples

A Rede Neural Simples é evidenciada como uma estrutura interligada por neurônios, que possuem a capacidade de realizar enormes processamentos de dados.(Grubler, 2018)


### Teste

```python
# Redimensionar os embeddings
reshaped_embeddings = np.reshape(embeddings, (1, -1, 1024))  # Redimensionar para (1, 12193, 1024)
reduced_embeddings = reshaped_embeddings[:, :2, :]  # Manter apenas as duas primeiras dimensões

# Reduzir a dimensão 1024 para 2
reduced_embeddings = tf.reduce_mean(reduced_embeddings, axis=2)  # Reduzir a dimensão 1024 para 2

# Definir os dados de treinamento
inputs = tf.constant(reduced_embeddings, dtype=tf.float32)
targets = tf.constant([pipeline_2], dtype=tf.float32)

# Definir a arquitetura da rede neural
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=2, activation='sigmoid', input_shape=(2,)),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
])

# Compilar o modelo
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['Accuracy',])


# Treinar a rede neural
model.fit(inputs, targets, epochs=1000)

# Testar a rede neural
predictions = model.predict(inputs)
print(predictions)
```
